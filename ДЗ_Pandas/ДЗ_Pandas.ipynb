# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
movies = pd.read_csv('movies.csv')
```

```python
ratings.head()
```

```python
movies.head()
```

```python
# Лучше сначала отфильтровать, потом соединять, так как join - очень ресурсоемкая операция

hight_score = 5.0

ratings = ratings[ ratings['rating'] == hight_score ].head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
# Я скачал маленькую базу, поэтому с оценками 5.0 получилось только 5 фильмов:

joined['title'].value_counts().head()
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power.info()
```

```python
# Страны прибалтики: Latvia, Lithuania, Estonia. Литву я нашел по фильтру:

power[ power['country'].str.contains('L', case=True) ]['country'].unique()
```

```python
# Сначала фильтрую по странам:

filtered_countries = power[ (power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia')]

filtered_countries.head()
```

```python
# Теперь по категориям:

filtered_countries = filtered_countries[ (filtered_countries['category'] == 4) | (filtered_countries['category'] == 12) | (filtered_countries['category'] == 21)]

filtered_countries.head()
```

```python
# Теперь по годам:

filtered_countries = filtered_countries[ (filtered_countries['year'] > 2004) & (filtered_countries['year'] < 2011) ]

filtered_countries.head()
```

```python
# Сортирую по убыванию, последние 3 позиции отрецательные:

filtered_countries.sort_values('quantity', ascending = False).tail()
```

```python
# Убираю отрецательные значения в quantity:

filtered_countries = filtered_countries[ (filtered_countries['quantity'] >= 0.0 )]
filtered_countries.sort_values('quantity', ascending = False).tail()
```
